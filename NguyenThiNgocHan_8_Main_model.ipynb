```python
# ============================================================
# KNN + LOGISTIC REGRESSION + SVM + PCA + LDA + K-MEANS
# CHIA DỮ LIỆU CHUẨN:
#   Bước 1: Tách TEST cố định trước
#   Bước 2: Phần còn lại mới chia TRAIN / VALIDATION theo 4:1, 7:3, 6:4
#
# KÝ HIỆU:
#   X_remain, y_remain : phần dữ liệu còn lại sau khi tách test
#   X_test, y_test     : tập test độc lập, giữ nguyên từ đầu
#   X_train, y_train   : tập huấn luyện
#   X_val, y_val       : tập validation
# ============================================================

import os
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.cluster import KMeans

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
    silhouette_score,
    davies_bouldin_score,
    adjusted_rand_score
)

from imblearn.over_sampling import SMOTE

os.makedirs("images", exist_ok=True)

# 1. LOAD DATA

zip_path = "cervical+cancer+risk+factors.zip"
extract_path = "./data"

if os.path.exists(zip_path):
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(extract_path)

    print("Giải nén xong!")
    print("Các file trong thư mục data:", os.listdir(extract_path))

    csv_path = os.path.join(extract_path, "risk_factors_cervical_cancer.csv")
else:
    csv_path = "./data/risk_factors_cervical_cancer.csv"

df = pd.read_csv(csv_path)

print("\nShape ban đầu:", df.shape)
display(df.head())


# 2. TIỀN XỬ LÝ DỮ LIỆU
# KIỂM TRA GIÁ TRỊ THIẾU VÀ GỢI Ý CÁCH ĐIỀN

# Thay '?' bằng NaN
df.replace("?", np.nan, inplace=True)

# Chuyển dữ liệu sang dạng số
df = df.apply(pd.to_numeric)

# 1. Thống kê số lượng và tỷ lệ giá trị thiếu

missing_count = df.isnull().sum()
missing_percent = df.isnull().mean() * 100

missing_df = pd.DataFrame({
    "Missing_Count": missing_count,
    "Missing_Percent": missing_percent
})

missing_df = missing_df[missing_df["Missing_Count"] > 0]
missing_df = missing_df.sort_values(by="Missing_Percent", ascending=False)

print("BẢNG THỐNG KÊ GIÁ TRỊ THIẾU:")
display(missing_df)


# 2. So sánh Mean - Median - Mode cho các cột bị thiếu

missing_cols = missing_df.index.tolist()

fill_compare = []

for col in missing_cols:
    mean_value = df[col].mean()
    median_value = df[col].median()
    mode_value = df[col].mode(dropna=True)

    if len(mode_value) > 0:
        mode_value = mode_value.iloc[0]
    else:
        mode_value = np.nan

    fill_compare.append({
        "Column": col,
        "Missing_Count": df[col].isnull().sum(),
        "Missing_Percent": df[col].isnull().mean() * 100,
        "Mean": mean_value,
        "Median": median_value,
        "Mode": mode_value,
        "Skewness": df[col].skew()
    })

fill_compare_df = pd.DataFrame(fill_compare)

print("\nBẢNG SO SÁNH MEAN - MEDIAN - MODE:")
display(fill_compare_df)


# 3. Gợi ý cách điền giá trị thiếu

def suggest_fill_method(series):
    """
    Gợi ý cách điền giá trị thiếu:
    - Nếu là biến nhị phân/category dạng 0/1: dùng mode
    - Nếu dữ liệu lệch nhiều: dùng median
    - Nếu dữ liệu tương đối đối xứng: dùng mean
    """
    non_missing = series.dropna()
    unique_values = non_missing.unique()

    # Biến nhị phân hoặc category mã hóa số
    if len(unique_values) <= 5:
        return "Mode"

    # Dữ liệu lệch nhiều
    skewness = non_missing.skew()

    if abs(skewness) > 1:
        return "Median"
    else:
        return "Mean"


fill_compare_df["Suggested_Method"] = fill_compare_df["Column"].apply(
    lambda col: suggest_fill_method(df[col])
)

print("\nGỢI Ý CÁCH ĐIỀN GIÁ TRỊ THIẾU:")
display(fill_compare_df[
    [
        "Column",
        "Missing_Count",
        "Missing_Percent",
        "Mean",
        "Median",
        "Mode",
        "Skewness",
        "Suggested_Method"
    ]
])





# Loại bỏ các cột có quá nhiều giá trị thiếu
cols_to_drop = [
    "STDs: Time since first diagnosis",
    "STDs: Time since last diagnosis"
]

df.drop(cols_to_drop, axis=1, inplace=True)

# Điền giá trị thiếu bằng median
df.fillna(df.median(), inplace=True)

print("\nShape sau tiền xử lý:", df.shape)
print("Số lượng missing còn lại:", df.isnull().sum().sum())

print("\nThông tin dữ liệu sau tiền xử lý:")
df.info()

print("\nThống kê mô tả:")
display(df.describe())


# 3. TÁCH BIẾN ĐẦU VÀO X VÀ BIẾN MỤC TIÊU y

TARGET_COL = "Biopsy"

X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL].astype(int)

print("\nKích thước X:", X.shape)
print("Kích thước y:", y.shape)

print("\nPhân bố nhãn ban đầu:")
print(y.value_counts().sort_index())

# 4. HÀM ĐÁNH GIÁ MÔ HÌNH

def get_classification_metrics(
    y_true,
    y_pred,
    model_name,
    data_type,
    split_name,
    eval_set
):
    """
    Trả về các chỉ số đánh giá mô hình phân loại.
    """
    return {
        "Train_Validation_Ratio": split_name,
        "Eval_Set": eval_set,
        "Model": model_name,
        "Data": data_type,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, zero_division=0),
        "Recall": recall_score(y_true, y_pred, zero_division=0),
        "F1-score": f1_score(y_true, y_pred, zero_division=0)
    }


def evaluate_model(y_true, y_pred, title):
    """
    In kết quả đánh giá mô hình.
    """
    print("\n" + "=" * 80)
    print(title)
    print("=" * 80)

    print("Accuracy :", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred, zero_division=0))
    print("Recall   :", recall_score(y_true, y_pred, zero_division=0))
    print("F1-score :", f1_score(y_true, y_pred, zero_division=0))

    print("\nConfusion Matrix:")
    print(confusion_matrix(y_true, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, zero_division=0))


def plot_confusion_matrix(y_true, y_pred, title, save_path):
    """
    Vẽ confusion matrix dạng heatmap.
    """
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(5, 4))

    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues"
    )

    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.savefig(save_path, dpi=300)
    plt.show()


def create_model(model_name, best_params):
    """
    Tạo lại mô hình từ bộ tham số tốt nhất.
    """
    if model_name == "KNN":
        return KNeighborsClassifier(**best_params)

    elif model_name == "Logistic Regression":
        return LogisticRegression(
            max_iter=3000,
            **best_params
        )

    elif model_name == "SVM":
        return SVC(**best_params)

    else:
        raise ValueError("Tên mô hình không hợp lệ.")


# 5. TÁCH TEST CỐ ĐỊNH TỪ ĐẦU

TEST_SIZE = 0.2

X_remain, X_test, y_remain, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=42,
    stratify=y
)

print("\n" + "#" * 80)
print("BƯỚC 1: TÁCH TEST CỐ ĐỊNH")
print("#" * 80)

print("Kích thước toàn bộ X :", X.shape)
print("Kích thước X_remain  :", X_remain.shape)
print("Kích thước X_test    :", X_test.shape)

print("\nPhân bố nhãn toàn bộ dữ liệu:")
print(y.value_counts().sort_index())

print("\nPhân bố nhãn y_remain:")
print(y_remain.value_counts().sort_index())

print("\nPhân bố nhãn y_test:")
print(y_test.value_counts().sort_index())


# 6. CHIA TRAIN / VALIDATION THEO 3 TỈ LỆ

# 4:1 nghĩa là validation chiếm 1/(4+1) = 0.2
# 7:3 nghĩa là validation chiếm 3/(7+3) = 0.3
# 6:4 nghĩa là validation chiếm 4/(6+4) = 0.4

split_configs = {
    "4:1": 0.2,
    "7:3": 0.3,
    "6:4": 0.4
}

all_results = []

best_model_info = {
    "best_recall": -1,
    "split_name": None,
    "model_name": None,
    "data_type": None,
    "best_params": None,
    "scaler": None,
    "pca": None,
    "lda": None
}


for split_name, val_size in split_configs.items():

    print("\n\n" + "#" * 80)
    print(f"BƯỚC 2: CHIA TRAIN / VALIDATION VỚI TỈ LỆ {split_name}")
    print("#" * 80)


    # 6.1 Chia X_remain thành train và validation

    X_train, X_val, y_train, y_val = train_test_split(
        X_remain,
        y_remain,
        test_size=val_size,
        random_state=42,
        stratify=y_remain
    )

    print("\nKích thước X_train:", X_train.shape)
    print("Kích thước X_val  :", X_val.shape)
    print("Kích thước X_test :", X_test.shape)

    print("\nPhân bố nhãn y_train:")
    print(y_train.value_counts().sort_index())

    print("\nPhân bố nhãn y_val:")
    print(y_val.value_counts().sort_index())

    print("\nPhân bố nhãn y_test:")
    print(y_test.value_counts().sort_index())

    # 6.2 Chuẩn hóa dữ liệu
    # Scaler chỉ fit trên X_train

    scaler = StandardScaler()

    X_train_scaled = scaler.fit_transform(X_train)
    X_val_scaled = scaler.transform(X_val)
    X_test_scaled = scaler.transform(X_test)

    # 6.3 Xử lý mất cân bằng bằng SMOTE
    # SMOTE chỉ áp dụng trên train

    smote = SMOTE(random_state=42)

    X_train_smote, y_train_smote = smote.fit_resample(
        X_train_scaled,
        y_train
    )

    print("\nPhân bố nhãn train trước SMOTE:")
    print(np.bincount(y_train.astype(int)))

    print("Phân bố nhãn train sau SMOTE:")
    print(np.bincount(y_train_smote.astype(int)))

    # 6.4 PCA giữ lại 95% phương sai
    # PCA chỉ fit trên train sau SMOTE

    pca = PCA(n_components=0.95, random_state=42)

    X_train_pca = pca.fit_transform(X_train_smote)
    X_val_pca = pca.transform(X_val_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    print("\nSố chiều sau PCA:", X_train_pca.shape[1])
    print("Tổng phương sai PCA giữ lại:", np.sum(pca.explained_variance_ratio_))

    # 6.5 LDA giảm về 1 chiều
    # Bài toán nhị phân nên LDA tối đa 1 chiều

    lda = LinearDiscriminantAnalysis(n_components=1)

    X_train_lda = lda.fit_transform(X_train_smote, y_train_smote)
    X_val_lda = lda.transform(X_val_scaled)
    X_test_lda = lda.transform(X_test_scaled)

    # 6.6 Tạo các dạng dữ liệu

    data_configs = {
        "Original": {
            "X_train_data": X_train_smote,
            "X_val_data": X_val_scaled,
            "X_test_data": X_test_scaled
        },
        "PCA": {
            "X_train_data": X_train_pca,
            "X_val_data": X_val_pca,
            "X_test_data": X_test_pca
        },
        "LDA": {
            "X_train_data": X_train_lda,
            "X_val_data": X_val_lda,
            "X_test_data": X_test_lda
        }
    }

    # 6.7 Grid Search riêng cho từng dạng dữ liệu

    for data_type, data_dict in data_configs.items():

        print("\n" + "-" * 80)
        print(f"GRID SEARCH TRÊN DỮ LIỆU: {data_type} - SPLIT {split_name}")
        print("-" * 80)

        X_train_data = data_dict["X_train_data"]
        X_val_data = data_dict["X_val_data"]
        X_test_data = data_dict["X_test_data"]

        # KNN
        param_knn = {
            "n_neighbors": np.arange(3, 21, 2)
        }

        grid_knn = GridSearchCV(
            KNeighborsClassifier(),
            param_knn,
            cv=5,
            scoring="recall"
        )

        grid_knn.fit(X_train_data, y_train_smote)


        # Logistic Regression

        param_lr = {
            "C": [0.01, 0.1, 1, 10],
            "solver": ["liblinear"]
        }

        grid_lr = GridSearchCV(
            LogisticRegression(max_iter=3000),
            param_lr,
            cv=5,
            scoring="recall"
        )

        grid_lr.fit(X_train_data, y_train_smote)

        # SVM

        param_svm = {
            "C": [0.1, 1, 10],
            "kernel": ["linear", "rbf"],
            "gamma": ["scale", "auto"]
        }

        grid_svm = GridSearchCV(
            SVC(),
            param_svm,
            cv=5,
            scoring="recall"
        )

        grid_svm.fit(X_train_data, y_train_smote)

        model_grids = {
            "KNN": grid_knn,
            "Logistic Regression": grid_lr,
            "SVM": grid_svm
        }

        print("Best KNN:", grid_knn.best_params_)
        print("Best Logistic Regression:", grid_lr.best_params_)
        print("Best SVM:", grid_svm.best_params_)

        # 6.8 Đánh giá trên Train, Validation, Test

        for model_name, grid_model in model_grids.items():

            best_params = grid_model.best_params_

            clf = create_model(model_name, best_params)

            clf.fit(X_train_data, y_train_smote)


            # Dự đoán trên TRAIN
            y_train_pred = clf.predict(X_train_data)

            result_train = get_classification_metrics(
                y_train_smote,
                y_train_pred,
                model_name,
                data_type,
                split_name,
                "Train"
            )

            all_results.append(result_train)

            # Dự đoán trên VALIDATION

            y_val_pred = clf.predict(X_val_data)

            evaluate_model(
                y_val,
                y_val_pred,
                f"{model_name} - {data_type} - Split {split_name} - Validation"
            )

            result_val = get_classification_metrics(
                y_val,
                y_val_pred,
                model_name,
                data_type,
                split_name,
                "Validation"
            )

            all_results.append(result_val)

            plot_confusion_matrix(
                y_val,
                y_val_pred,
                title=f"Confusion Matrix - {model_name} - {data_type} - {split_name} - Validation",
                save_path=f"images/cm_{model_name}_{data_type}_{split_name}_validation.png".replace(" ", "_").replace(":", "_")
            )

            # Dự đoán trên TEST

            y_test_pred = clf.predict(X_test_data)

            evaluate_model(
                y_test,
                y_test_pred,
                f"{model_name} - {data_type} - Split {split_name} - Test"
            )

            result_test = get_classification_metrics(
                y_test,
                y_test_pred,
                model_name,
                data_type,
                split_name,
                "Test"
            )

            all_results.append(result_test)

            # Lưu mô hình tốt nhất theo Recall trên Validation

            val_recall = recall_score(y_val, y_val_pred, zero_division=0)

            if val_recall > best_model_info["best_recall"]:

                best_model_info["best_recall"] = val_recall
                best_model_info["split_name"] = split_name
                best_model_info["model_name"] = model_name
                best_model_info["data_type"] = data_type
                best_model_info["best_params"] = best_params
                best_model_info["scaler"] = scaler
                best_model_info["pca"] = pca
                best_model_info["lda"] = lda


# 7. BẢNG TỔNG HỢP KẾT QUẢ PHÂN LOẠI
results_df = pd.DataFrame(all_results)

print("\n\nBẢNG TỔNG HỢP KẾT QUẢ PHÂN LOẠI:")
display(results_df)

results_df.to_csv(
    "classification_results_train_validation_test.csv",
    index=False
)

print("Đã lưu file: classification_results_train_validation_test.csv")


# 8. KIỂM TRA OVERFITTING

# So sánh Train và Validation theo F1-score, Recall, Accuracy

train_results = results_df[results_df["Eval_Set"] == "Train"]
val_results = results_df[results_df["Eval_Set"] == "Validation"]

overfit_df = pd.merge(
    train_results,
    val_results,
    on=["Train_Validation_Ratio", "Model", "Data"],
    suffixes=("_Train", "_Validation")
)

overfit_df["Accuracy_Diff"] = overfit_df["Accuracy_Train"] - overfit_df["Accuracy_Validation"]
overfit_df["Recall_Diff"] = overfit_df["Recall_Train"] - overfit_df["Recall_Validation"]
overfit_df["F1_Diff"] = overfit_df["F1-score_Train"] - overfit_df["F1-score_Validation"]

print("\n\nBẢNG KIỂM TRA OVERFITTING:")
display(
    overfit_df[
        [
            "Train_Validation_Ratio",
            "Model",
            "Data",
            "Accuracy_Train",
            "Accuracy_Validation",
            "Accuracy_Diff",
            "Recall_Train",
            "Recall_Validation",
            "Recall_Diff",
            "F1-score_Train",
            "F1-score_Validation",
            "F1_Diff"
        ]
    ]
)

overfit_df.to_csv(
    "overfitting_check_results.csv",
    index=False
)

print("Đã lưu file: overfitting_check_results.csv")


# 9. BIỂU ĐỒ SO SÁNH RECALL

plt.figure(figsize=(14, 6))

sns.barplot(
    data=results_df[results_df["Eval_Set"] == "Validation"],
    x="Model",
    y="Recall",
    hue="Data"
)

plt.title("So sánh Recall trên tập Validation")
plt.ylabel("Recall")
plt.xlabel("Model")
plt.tight_layout()

plt.savefig(
    "images/recall_validation_comparison.png",
    dpi=300
)

plt.show()


# 10. BIỂU ĐỒ SO SÁNH ACCURACY THEO TỈ LỆ TRAIN / VALIDATION

plt.figure(figsize=(14, 6))

sns.barplot(
    data=results_df[results_df["Eval_Set"] == "Validation"],
    x="Train_Validation_Ratio",
    y="Accuracy",
    hue="Model"
)

plt.title("So sánh Accuracy theo các tỷ lệ Train/Validation")
plt.ylabel("Accuracy")
plt.xlabel("Train:Validation")
plt.tight_layout()

plt.savefig(
    "images/accuracy_by_train_validation_ratio.png",
    dpi=300
)

plt.show()


# 11. BIỂU ĐỒ SO SÁNH VALIDATION VÀ TEST

val_test_results = results_df[
    results_df["Eval_Set"].isin(["Validation", "Test"])
]

plt.figure(figsize=(14, 6))

sns.barplot(
    data=val_test_results,
    x="Train_Validation_Ratio",
    y="F1-score",
    hue="Eval_Set"
)

plt.title("So sánh F1-score giữa Validation và Test")
plt.ylabel("F1-score")
plt.xlabel("Train:Validation")
plt.tight_layout()

plt.savefig(
    "images/f1_validation_test_comparison.png",
    dpi=300
)

plt.show()


# 12. TRỰC QUAN HÓA PCA 2D

pca_vis = PCA(n_components=2, random_state=42)

X_pca_vis = pca_vis.fit_transform(X_train_smote)

pca_df = pd.DataFrame({
    "PC1": X_pca_vis[:, 0],
    "PC2": X_pca_vis[:, 1],
    "Class": y_train_smote
})

plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="Class",
    palette="Set1",
    s=70,
    alpha=0.8
)

plt.title("PCA - Biểu diễn dữ liệu trên 2 thành phần chính đầu tiên")

plt.xlabel(
    f"PC1 ({pca_vis.explained_variance_ratio_[0] * 100:.2f}% variance)"
)

plt.ylabel(
    f"PC2 ({pca_vis.explained_variance_ratio_[1] * 100:.2f}% variance)"
)

plt.grid(True)
plt.tight_layout()

plt.savefig(
    "images/pca_2d_visualization.png",
    dpi=300
)

plt.show()


# 13. TRỰC QUAN HÓA LDA

lda_vis = LinearDiscriminantAnalysis(n_components=1)

X_lda_vis = lda_vis.fit_transform(X_train_smote, y_train_smote)

lda_df = pd.DataFrame({
    "LD1": X_lda_vis[:, 0],
    "Class": y_train_smote
})

plt.figure(figsize=(8, 4))

sns.scatterplot(
    data=lda_df,
    x="LD1",
    y=np.zeros(len(lda_df)),
    hue="Class",
    palette="Set1",
    s=70,
    alpha=0.8
)

plt.title("LDA - Biểu diễn dữ liệu sau giảm chiều")
plt.xlabel("LD1")
plt.yticks([])
plt.grid(True)
plt.tight_layout()

plt.savefig(
    "images/lda_visualization.png",
    dpi=300
)

plt.show()


# 14. EXPLAINED VARIANCE CỦA PCA

pca_full = PCA(random_state=42)
pca_full.fit(X_train_smote)

explained = pca_full.explained_variance_ratio_
cum_var = np.cumsum(explained)

pca_variance_df = pd.DataFrame({
    "Principal_Component": [f"PC{i+1}" for i in range(len(explained))],
    "Explained_Variance_Ratio": explained,
    "Cumulative_Explained_Variance": cum_var
})

print("\nBảng explained variance của PCA:")
display(pca_variance_df.head(10))

pca_variance_df.to_csv(
    "pca_explained_variance.csv",
    index=False
)

plt.figure(figsize=(10, 6))

plt.bar(
    range(1, 7),
    explained[:6]
)

plt.plot(
    range(1, 7),
    explained[:6],
    marker="o"
)

plt.xlabel("Thành phần chính")
plt.ylabel("Tỷ lệ phương sai giải thích")
plt.title("Tỷ lệ phương sai của 6 thành phần chính đầu tiên")
plt.xticks(range(1, 7))
plt.grid(True)
plt.tight_layout()

plt.savefig(
    "images/pca_explained_variance.png",
    dpi=300
)

plt.show()


plt.figure(figsize=(10, 6))

plt.plot(
    range(1, 7),
    cum_var[:6],
    marker="o"
)

plt.xlabel("Số thành phần chính")
plt.ylabel("Phương sai tích lũy")
plt.title("Phương sai tích lũy của PCA")
plt.xticks(range(1, 7))
plt.grid(True)
plt.tight_layout()

plt.savefig(
    "images/pca_cumulative_variance.png",
    dpi=300
)

plt.show()


# 15. TRỰC QUAN THEO CÁC CẶP THÀNH PHẦN CHÍNH

pca_pair = PCA(n_components=6, random_state=42)

X_pca_pair = pca_pair.fit_transform(X_train_smote)

pairs = [
    (0, 1),
    (0, 2),
    (1, 2),
    (2, 3),
    (3, 4),
    (4, 5)
]

titles = [
    "PC1 vs PC2",
    "PC1 vs PC3",
    "PC2 vs PC3",
    "PC3 vs PC4",
    "PC4 vs PC5",
    "PC5 vs PC6"
]

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for idx, ((i, j), title) in enumerate(zip(pairs, titles)):

    ax = axes[idx]

    for label in np.unique(y_train_smote):

        ax.scatter(
            X_pca_pair[y_train_smote == label, i],
            X_pca_pair[y_train_smote == label, j],
            label=f"Class {label}",
            alpha=0.7
        )

    ax.set_title(title)
    ax.set_xlabel(f"PC{i + 1}")
    ax.set_ylabel(f"PC{j + 1}")
    ax.grid(True)

handles, labels = axes[0].get_legend_handles_labels()

fig.legend(
    handles,
    labels,
    loc="upper right"
)

plt.suptitle(
    "Phân tán dữ liệu theo các cặp thành phần chính",
    fontsize=16
)

plt.tight_layout()

plt.savefig(
    "images/pca_pair_scatter.png",
    dpi=300
)

plt.show()


# 16. PHÂN CỤM DỮ LIỆU BẰNG K-MEANS

print("\n\n" + "#" * 80)
print("PHÂN CỤM DỮ LIỆU BẰNG K-MEANS")
print("#" * 80)

# Phân cụm bỏ qua nhãn y
scaler_cluster = StandardScaler()
X_cluster_scaled = scaler_cluster.fit_transform(X)

# 16.1 Elbow Method
wcss = []
k_values = range(2, 11)

for k in k_values:

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    kmeans.fit(X_cluster_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))

plt.plot(
    list(k_values),
    wcss,
    marker="o"
)

plt.xlabel("Số cụm K")
plt.ylabel("WCSS / Inertia")
plt.title("Elbow Method for K-Means")
plt.grid(True)
plt.tight_layout()

plt.savefig(
    "images/kmeans_elbow.png",
    dpi=300
)

plt.show()


# 16.2 Silhouette Score
silhouette_scores = []

for k in k_values:

    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    cluster_labels_temp = kmeans.fit_predict(X_cluster_scaled)

    score = silhouette_score(
        X_cluster_scaled,
        cluster_labels_temp
    )

    silhouette_scores.append(score)

plt.figure(figsize=(8, 5))

plt.plot(
    list(k_values),
    silhouette_scores,
    marker="o"
)

plt.xlabel("Số cụm K")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for K-Means")
plt.grid(True)
plt.tight_layout()

plt.savefig(
    "images/kmeans_silhouette.png",
    dpi=300
)

plt.show()


# 16.3 Chọn K = 2 để so sánh với hai nhãn Biopsy

best_k = 2

kmeans_final = KMeans(
    n_clusters=best_k,
    random_state=42,
    n_init=10
)

cluster_labels = kmeans_final.fit_predict(X_cluster_scaled)


# 16.4 Đánh giá phân cụm
sil_score = silhouette_score(
    X_cluster_scaled,
    cluster_labels
)

db_index = davies_bouldin_score(
    X_cluster_scaled,
    cluster_labels
)

ari_score = adjusted_rand_score(
    y,
    cluster_labels
)

print("\nSilhouette Score:", sil_score)
print("Davies-Bouldin Index:", db_index)
print("Adjusted Rand Index so với nhãn Biopsy:", ari_score)

cluster_table = pd.crosstab(
    pd.Series(cluster_labels, name="Cluster"),
    pd.Series(y.values, name="Biopsy")
)

print("\nBảng so sánh Cluster với nhãn thật Biopsy:")
display(cluster_table)


# 16.5 Trực quan hóa cụm bằng PCA 2D

pca_cluster = PCA(n_components=2, random_state=42)

X_cluster_pca = pca_cluster.fit_transform(X_cluster_scaled)

cluster_vis_df = pd.DataFrame({
    "PC1": X_cluster_pca[:, 0],
    "PC2": X_cluster_pca[:, 1],
    "Cluster": cluster_labels,
    "Biopsy": y.values
})

plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=cluster_vis_df,
    x="PC1",
    y="PC2",
    hue="Cluster",
    style="Biopsy",
    palette="Set1",
    s=70,
    alpha=0.8
)

plt.title("K-Means Clustering Visualization on PCA 2D")
plt.xlabel("PC1")
plt.ylabel("PC2")

plt.tight_layout()

plt.savefig(
    "images/kmeans_cluster_pca_2d.png",
    dpi=300
)

plt.show()


# 16.6 Lưu kết quả phân cụm

cluster_result_df = X.copy()
cluster_result_df["Biopsy"] = y.values
cluster_result_df["KMeans_Cluster"] = cluster_labels

cluster_result_df.to_csv(
    "kmeans_cluster_results.csv",
    index=False
)

print("Đã lưu kết quả phân cụm vào file: kmeans_cluster_results.csv")


# 17. IN MÔ HÌNH TỐT NHẤT THEO RECALL TRÊN VALIDATION

print("\n\n" + "#" * 80)
print("MÔ HÌNH TỐT NHẤT THEO RECALL TRÊN VALIDATION")
print("#" * 80)

print("Tỷ lệ Train:Validation:", best_model_info["split_name"])
print("Mô hình:", best_model_info["model_name"])
print("Dữ liệu:", best_model_info["data_type"])
print("Best params:", best_model_info["best_params"])
print("Recall Validation tốt nhất:", best_model_info["best_recall"])
```

# Chuyển đổi bài toán phân loại sang bài toán hồi quy

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.decomposition import PCA
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

# 1. Tạo đầu ra hồi quy mới từ Logistic Regression
clf = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
clf.fit(X_train_scaled, y_train)

# y_reg là xác suất thuộc lớp 1: Biopsy = 1
y_train_reg = clf.predict_proba(X_train_scaled)[:, 1]
y_test_reg = clf.predict_proba(X_test_scaled)[:, 1]


# 2. Hàm đánh giá hồi quy
def evaluate_regression(model, X_train, X_test, y_train_reg, y_test_reg, model_name, data_type):
    model.fit(X_train, y_train_reg)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test_reg, y_pred)
    mse = mean_squared_error(y_test_reg, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test_reg, y_pred)

    return {
        "Mô hình": model_name,
        "Dữ liệu": data_type,
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "R2": r2
    }

results_reg = []

# 3. Hồi quy trên dữ liệu gốc
models_reg = {
    "Linear Regression": LinearRegression(),
    "KNN Regressor": KNeighborsRegressor(n_neighbors=5)
}

for name, model in models_reg.items():
    results_reg.append(
        evaluate_regression(
            model,
            X_train_scaled,
            X_test_scaled,
            y_train_reg,
            y_test_reg,
            name,
            "Original"
        )
    )

# 4. Hồi quy trên dữ liệu giảm còn 1/3 số chiều bằng PCA
n_components = X_train_scaled.shape[1] // 3

pca_13 = PCA(n_components=n_components, random_state=42)
X_train_pca_13 = pca_13.fit_transform(X_train_scaled)
X_test_pca_13 = pca_13.transform(X_test_scaled)

for name, model in models_reg.items():
    results_reg.append(
        evaluate_regression(
            model,
            X_train_pca_13,
            X_test_pca_13,
            y_train_reg,
            y_test_reg,
            name,
            "PCA 1/3 dimensions"
        )
    )


# 5. In bảng kết quả
results_reg_df = pd.DataFrame(results_reg)
print(results_reg_df)
```
